```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# SEBASTI√ÅN ESC√ÅRPITA S√ÅNCHEZ
# Establecer seed
np.random.seed(23041008)

# Generar 100 n√∫meros aleatorios entre 0 y 1
numeros_aleatorios = np.random.uniform(0, 1, 100)

# Realizar la prueba de uniformidad de Kolmogorov-Smirnov
# H0: Los datos siguen una distribuci√≥n uniforme
# H1: Los datos no siguen una distribuci√≥n uniforme
estadistico_ks, valor_p = stats.kstest(numeros_aleatorios, 'uniform')

# Imprimir resultados
print("Resultados de la prueba de uniformidad de Kolmogorov-Smirnov:")
print(f"Estad√≠stico KS: {estadistico_ks:.4f}")
print(f"Valor p: {valor_p:.4f}")

# Visualizar los resultados
plt.figure(figsize=(10, 5))

# Histograma de los n√∫meros aleatorios
plt.subplot(1, 2, 1)
plt.hist(numeros_aleatorios, bins=10, edgecolor='black')
plt.title('Histograma de los N√∫meros Aleatorios')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

# Gr√°fico de la Funci√≥n de Distribuci√≥n Acumulativa (CDF)
plt.subplot(1, 2, 2)
datos_ordenados = np.sort(numeros_aleatorios)
y = np.arange(1, len(datos_ordenados) + 1) / len(datos_ordenados)
plt.step(datos_ordenados, y, label='CDF Emp√≠rica')
plt.plot(datos_ordenados, datos_ordenados, 'r--', label='CDF Uniforme')
plt.title('CDF Emp√≠rica vs CDF Uniforme')
plt.xlabel('Valor')
plt.ylabel('Probabilidad Acumulada')
plt.legend()

plt.tight_layout()
plt.show()

# Interpretaci√≥n de los resultados
nivel_significancia = 0.05  # Nivel de significancia
print("\nInterpretaci√≥n:")
if valor_p > nivel_significancia:
    print(f"No se rechaza H0 (valor p = {valor_p:.4f} > {nivel_significancia})")
    print("Los datos parecen estar distribuidos uniformemente.")
else:
    print(f"Se rechaza H0 (valor p = {valor_p:.4f} <= {nivel_significancia})")
    print("Los datos no parecen estar distribuidos uniformemente.")
```
